```python
from itertools import permutations

# Players
players = ["Sanpi", "Preyasi", "Jhanvi", "Niels", "Aiden", "Benjamin", "Youssra"]

# Possible targets for each player
possible_targets = {
    "Sanpi": ["Preyasi", "Jhanvi", "Youssra"],
    "Preyasi": ["Niels", "Aiden"],
    "Jhanvi": ["Sanpi", "Aiden", "Niels", "Benjamin"],
    "Niels": ["Preyasi", "Jhanvi", "Aiden", "Benjamin", "Youssra"],
    "Aiden": ["Preyasi", "Jhanvi", "Niels", "Youssra"],
    "Benjamin": ["Preyasi", "Jhanvi", "Niels", "Aiden"],
    "Youssra": ["Sanpi", "Niels", "Aiden"]
}

# Check if assignment is valid
def is_valid_assignment(assignment):
    assigned_targets = set()
    for p, t in assignment.items():
        # Check if each target is allowed
        if t not in possible_targets[p]:
            return False
        # Check for duplicate targets
        if t in assigned_targets:
            return False
        assigned_targets.add(t)
        # Check for mutual assignment
        if t in assignment and assignment[t] == p:
            return False
    return True

# Generate all valid assignments
valid_assignments = []
for target_permutation in permutations(players):
    assignment = dict(zip(players, target_permutation))
    if is_valid_assignment(assignment):
        valid_assignments.append(assignment)

# Print the first 16 valid assignments
for i, va in enumerate(valid_assignments[:16], 1):
    print(f"Option {i}:")
    for player, target in va.items():
        print(f"  {player} -> {target}")
    print()

print(f"Total valid assignments: {len(valid_assignments)}")
```
